```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  row, col = floor.shape

  for i in range(col):
    if i%2 == 0:
      for j in range(0, row, 2):
        print(floor[j][i],end = ' ')
    else:
      if row%2 == 0:
        for k in range(row-1, -1, -2):
          print(floor[k][i],end = ' ')
      else:
        for k in range(row-2, -1, -2):
          print(floor[k][i],end = ' ')
    print()




floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    new_arr = np.full((district_row + 2*depth,district_col + 2*depth), 8, dtype = int)
    for i in range(district_row):
      for j in range(district_col):
        new_arr[i+depth][j+depth] = district[i][j]
    return new_arr
depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)

#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  row, col = clubroom.shape
  size = int((row*col - row) / 2)
  strength_diff = np.zeros(size, dtype = int)
  k = 0
  for i in range(row):
    for j in range(col):
      if i != j and j > i:
        strength_diff[k] = clubroom[i][j] - clubroom[j][i]
        k += 1
  return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  matx = np.array(([[None]*n]*n))
  for i in range(1, n+1):
    for j in range(i):
      if j == 0 or j == i-1:
        matx[i-1][j] = i
      else:
        summ = 0
        for k in range(1, j+2):
          if matx[i-2][k-1] == None:
            summ+=0
          else:
            summ+= matx[i-2][k-1]
        matx[i-1][j] = summ
  return matx

def print_without_none(matrix):
  r, c = matrix.shape
  for i in range(r):
    for j in range(c):
      if matrix[i][j] != None:
        print(matrix[i][j], end = '  ')
    print()



n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT

```python
def moving_around(cmds):
  grid = np.full((7, 7), '.')
  r = 3
  c = 3
  grid[r][c] = '-'
  for i in range(len(cmds)):
    if i != 0:
      grid[r][c] = '*'
    if cmds[i] == 1:
      if -1<r-2<7 and -1<c-3<7:
        r -= 2
        c -= 3
    if cmds[i] == 2:
      if -1<r-2<7 and -1<c-1<7:
        r -= 2
        c -= 1
    if cmds[i] == 3:
      if -1<r-3<7 and -1<c-2<7:
        r -= 3
        c -= 2
    if cmds[i] == 4:
      if -1<r-1<7 and -1<c-2<7:
        r -= 1
        c -= 2
    if cmds[i] == 5:
      if -1<r-2<7 and -1<c+1<7:
        r -= 2
        c += 1
    if cmds[i] == 6:
      if -1<r-2<7 and -1<c+3<7:
        r -= 2
        c += 3
    if cmds[i] == 7:
      if -1<r-3<7 and -1<c+2<7:
        r -= 3
        c += 2
    if cmds[i] == 8:
      if -1<r-1<7 and -1<c+2<7:
        r -= 1
        c += 2
    if cmds[i] == 9:
      if -1<r+2<7 and -1<c-3<7:
        r += 2
        c -= 3
    if cmds[i] == 10:
      if -1<r+2<7 and -1<c-1<7:
        r += 2
        c -= 1
    if cmds[i] == 11:
      if -1<r+1<7 and -1<c-2<7:
        r += 1
        c -= 2
    if cmds[i] == 12:
      if -1<r+3<7 and -1<c-2<7:
        r += 3
        c -= 2
    grid[r][c] = '/'
  return grid


cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
