```python
class Patient:
  def __init__(self, id, name, age, bloodgroup):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = None
    self.prev = None

class WRM:
  def __init__(self):
    self.head = Patient(None, None, None, None)
    self.head.next = self.head
    self.head.prev = self.head
    self.old_patient = self.head

  def register_patient(self, id, name, age, bloodgroup):
    new_patient = Patient(id, name, age, bloodgroup)
    self.old_patient.next = new_patient
    new_patient.next = self.head
    self.head.prev = new_patient
    new_patient.prev = self.old_patient
    self.old_patient = new_patient
    print("Patient added successfully!")

  def serve_patient(self):
    if self.head.next != self.head:
      serve = self.head.next
      self.head.next = serve.next
      serve.next.prev = self.head
      serve.next = None
      serve.prev = None
      print(serve.name, 'is served')
    else:
      print('All patient are served')

  def cancel_all(self):
    self.head.next = self.head
    self.head.prev = self.head
    print('All appointments are cancelled')

  def can_doctor_go_home(self):
    if self.head == self.head.next:
      print('Yes')
    else:
      print('No')

  def show_all_patients(self):
    temp = self.head.next
    if temp == self.head:
      print('No patient is remaining')
    while temp != self.head:
      if temp.next == self.head:
        print(temp.name)
      else:
        print(temp.name, end =', ')
      temp = temp.next

  def reverse_the_line(self):
    if self.head == self.head.next:
      print('No patients in the waiting room.')
    else:
      cur_node = self.head
      while True:
        temp1 = cur_node.prev
        temp2 = cur_node.next
        cur_node.next = temp1
        cur_node.prev = temp2
        cur_node = cur_node.prev
        if cur_node == self.head:
          break
      print("Line reversed successfully.")
```

```python
import numpy as np

class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p

class DoublyList:

  # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
  def __init__(self):
    self.head = Node(None, None, None)
    self.head.next = self.head
    self.head.prev = self.head


  #create a DHDCL from an array
  def create_list(self, l):
    tail = self.head
    for i in range(len(l)):
      n = Node(l[i], None, tail )
      tail.next = n
      tail = n
    tail.next = self.head
    self.head.prev = tail


  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    i = self.head.next
    count = 0
    while i != self.head:
      count += 1
      i = i.next
    return count


  # prints the elements in the list forward
  def forwardprint(self):
    i = self.head.next
    while i != self.head:
      print(i.element)
      i = i.next


  # prints the elements in the list backward
  def backwardprint(self):
    i = self.head.prev
    while i != self.head:
      print(i.element)
      i = i.prev


  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    i = self.head.next
    count = 0
    while i != self.head:
      if count == idx:
        return i
      count += 1
      i = i.next
    return None

  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    i = self.head.next
    count = 0
    while i != self.head:
      if i.element == elem:
        return count
      count += 1
      i = i.next
    return -1


  # insert at specific index of the DHDCL
  def insert(self, elem, idx):
    count = -1
    i = self.head
    while True:
      if count == idx-1:
        n = Node(elem, None, None)
        n.next = i.next
        n.prev = i
        i.next = n
        i.next.next.prev = n
        break
      else:
        i = i.next
        count += 1


  #removes the node of the specific index and returns the element of the node
  def remove(self, idx):
    count = -1
    i = self.head
    while True:
      if count == idx-1:
        temp = i.next
        i.next = i.next.next
        i.next.prev = i
        temp.next = None
        temp.prev = None
        return temp.element
      else:
        i = i.next
        count += 1
```
